```python
import plotly.offline as pyo    
import plotly.graph_objs as go  
import pandas as pd 
import sys 

sys.path.append('../')

import omdutils as omd 

pd.options.display.max_columns = 999
```

```python
sql = 'SELECT * FROM cars'
db = '../SourceData/datasets.db'
df = omd.from_sqlite(sql, db)
df.head()
```

```python
df.info()
```

```python
# data['origin'] = data['origin'].replace({1: 'USA', 2: 'Europe', 3: 'Japan'})

data = [go.Scatter(x=df['model_year'], 
                   y=df['horsepower'], 
                   mode='markers', 
                   text=df['name'].str.upper(),
                   marker=dict(size=df['mpg'],
                               color=df['origin'],
                               showscale=True,
                               colorscale='Jet',
                               colorbar=dict(title='Origin'),))]

layout = go.Layout(title='Horsepower vs Model Year', 
                   title_x=0.5,
                   titlefont=dict(size=24),
                   xaxis_title='Model Year', 
                   yaxis_title='Horsepower',
                   hovermode='closest',
                   height=750,
                   annotations=[dict(
                                    x=0.5,  # Horizontal center
                                    y=1.08,  # Slightly above the top of the plot
                                    xref='paper',
                                    yref='paper',
                                    text='(mpg determines bubble size)',
                                    showarrow=False,
                                    font=dict(size=20)
                                )
                               ])

fig = go.Figure(data=data, layout=layout)
fig.show()
```
