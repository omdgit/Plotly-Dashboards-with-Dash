```python
#######
# This script creates a static matplotlib plot
######
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# create fake data:
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
df.plot()
plt.show()
#######
# At the terminal run:  python basic1.py
# Close the plot window to close the script
######
```

```python
#######
# This script creates the same type of plot as basic1.py,
# but in Plotly. Note that it creates an .html file!
######
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go

# create fake data:
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
display(df.head())
pyo.plot([{
    'x': df.index,
    'y': df[col],
    'name': col
} for col in df.columns])
```

## Scatter Plot

```python
df['E'] = df['A'] + np.random.randint(-2,2,100) * 0.1

# Assuming df is already defined with columns 'A' and 'E'
data = [go.Scatter(x=df['A'], 
                   y=df['E'], 
                   mode='markers',
                   marker=dict(
                       size=12,
                       color='rgb(51,204,153)',
                       symbol='pentagon',
                       line={'width':2}
                   ))]

layout = go.Layout(
    title='Scatterplot',
    title_x=0.5,  # Center the title
    titlefont={'size':28},
    xaxis={'title':'Feature A', 'titlefont':{'size':22}}, 
    yaxis={'title':'Feature E', 'titlefont':{'size':22}},
    hovermode='closest',
    annotations=[
        dict(
            x=0.5,  # Horizontal center
            y=1.05,  # Slightly above the top of the plot
            xref='paper',
            yref='paper',
            text='Feature "A" vs Feature "E"',
            showarrow=False,
            font=dict(size=20)
        )
    ]
)

fig = go.Figure(data=data, layout=layout)
pyo.plot(fig, filename='basic2.html')
```
