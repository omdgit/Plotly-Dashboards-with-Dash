# Box Plots

```python
import plotly.offline as pyo 
import plotly.graph_objs as go 
import pandas as pd 
import numpy as np
import sys 

sys.path.append('../') 

import omdutils as omd 

pd.options.display.max_columns = 999
```

```python
#######
# Objective: Make a DataFrame using the Abalone dataset (../data/abalone.csv).
# Take two independent random samples of different sizes from the 'rings' field.
# HINT: np.random.choice(df['rings'],10,replace=False) takes 10 random values
# Use box plots to show that the samples do derive from the same population.
######
# df = pd.read_csv('../Data/abalone.csv')
db = '../SourceData/datasets.db'
sql = 'SELECT * FROM abalone'
# omd.to_sqlite(df, 'abalone', db)
df = omd.from_sqlite(sql, db)
df
```

```python
number_of_samples = 10
samples = [f'Sample {i+1}' for i in range(number_of_samples)]

data = [go.Box(y=np.random.choice(df['rings'], 250, replace=False),
               name=sample,
               marker=dict(size=2)) for sample in samples]

layout = go.Layout(title=f'Comparison of {number_of_samples} samples',
                   title_x=0.5,
                   titlefont=dict(size=20))
fig = go.Figure(data=data, layout=layout)
fig.show()
```
