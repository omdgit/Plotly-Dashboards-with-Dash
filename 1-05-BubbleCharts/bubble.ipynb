# Bubble Charts

```python
import plotly.offline as pyo   
import plotly.graph_objs as go 
import pandas as pd
import sys 

sys.path.append('../')

import omdutils as omd  

pd.options.display.max_columns = 999
```

```python
# df = pd.read_csv('../Data/mpg.csv')
# df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
# omd.to_sqlite(df, 'cars', '../SourceData/datasets.db')
df = omd.from_sqlite('SELECT * FROM cars', '../SourceData/datasets.db')
df.head()
```

## Bubble Chart

```python
data = [go.Scatter(x=df['mpg'],
                   y=df['horsepower'],
                   mode='markers',
                   marker=dict(
                       size=df['weight']/100,
                       color=df['cylinders'],
                       symbol='circle',
                       line={'width':2},
                       colorscale='Jet',
                       showscale=True,
                       colorbar=dict(title='Cylinders')),
                   text=df['name'].str.upper()),
                   ] 

layout = go.Layout(title='MPG vs Horsepower',
                   title_x=0.5,
                   height=600,
                   titlefont=dict(size=24),
                   xaxis=dict(title='MPG'),
                   yaxis=dict(title='Horsepower'),
                   legend=dict(title='Weight (normalized)'),
                   annotations=[dict(
                                    x=0.5,  # Horizontal center
                                    y=1.10,  # Slightly above the top of the plot
                                    xref='paper',
                                    yref='paper',
                                    text='(Weight determines bubble size)',
                                    showarrow=False,
                                    font=dict(size=20)
                                )
                               ]
                   )

fig = go.Figure(data=data, layout=layout)
pyo.plot(fig, filename='bubblechart.html')
fig.show()
```
